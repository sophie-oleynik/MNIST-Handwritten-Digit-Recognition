```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
from keras import optimizers
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from keras import regularizers
```

Зчитуємо дані і відображаємо train.csv:

```
df = pd.read_csv('/content/drive/My Drive/root/train.csv')
test = pd.read_csv('/content/drive/My Drive/root/test.csv')
df.head()
```

Розділяємо train.csv для тренування системи:

```
df_train, df_test = train_test_split(df, stratify=df['label'], random_state=42)
df_test.head()
```

Форматування і нормалізація даних:

```
x_train = df_train.values[:,2:]
x_test = df_test.values[:,2:]
y_train = df_train.values[:,1]
y_test = df_test.values[:,1]

input_shape = (28, 28, 1)

x_train.shape
x_test.shape

x_train = x_train.reshape(x_train.shape[0], 28, 28, 1)
x_train = x_train.astype('float32')
x_train /= 255
x_test = x_test.reshape(x_test.shape[0], 28, 28, 1)
x_test = x_test.astype('float32')
x_test /= 255
```

Побудова згорткової нейронної мережі:

```
# Імпорт необхідних модулів Keras, що містять модель та шари
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, Dropout, Flatten, MaxPooling2D
# Створення послідовної моделі та додавання шарів
model = Sequential()
model.add(Conv2D(28, kernel_size=(3,3), input_shape=input_shape))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten()) # Flattening the 2D arrays for fully connected layers
model.add(Dense(128, activation=tf.nn.relu))
model.add(Dropout(0.2))
model.add(Dense(10,activation=tf.nn.softmax))
```

Компіляція моделі:

```
model.compile(optimizer='adam', 
              loss='sparse_categorical_crossentropy', 
              metrics=['accuracy'])
model.fit(x=x_train,y=y_train, epochs=20)
```

```
model.evaluate(x_test, y_test)
```

Форматування і нормалізація даних для тесту:

```
x_mytest = test.values[:,1:]
x_mytest.shape
x_mytest = x_mytest.reshape(x_mytest.shape[0], 28, 28, 1)
x_mytest = x_mytest.astype('float32')
x_mytest /= 255
```

Результат:


```
predictions = model.predict(x_mytest)
```

```
rounded = [np.argmax(i) for i in predictions]
print(rounded)
```
